```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import make_circles
```

```python
circles = make_circles(n_samples=1000, noise=.15, factor=.1, random_state=42)[0]
circles[:5]
```

```python
# Instantiate Standard Scaler
scaler = StandardScaler()
# Fit & transform data.
circles = scaler.fit_transform(circles)
plt.scatter(circles[:,0], circles[:,1])
```

```python
# BASELINE DBSCAN
```

```python
# Instantiate Standard Scaler
scaler = StandardScaler()
# Fit & transform data.
circles = scaler.fit_transform(circles)
plt.scatter(circles[:,0], circles[:,1])
```

```python
# TUNING DBSCAN
```

```python
# choose min_samples and tune epsilon

min_samples = 20
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors.fit(circles)
distances, indices = neighbors.kneighbors(circles)
distances[:5]
```

```python
# Obtain the furthest distance for each point
farthest_distances = distances[:, -1]
farthest_distances
```

```python
# Sort the furthest distance for each point
sorted_distances = np.sort(farthest_distances)
sorted_distances[0:5]
```

```python
# Plot the distances required for points to be included in a cluster with min sample of 20
plt.plot(sorted_distances)
plt.xlabel("Number of Points")
plt.ylabel("Distance to 19th Neighbor")
plt.show()
```

```python
# FIT THE TUNED DBSCAN

# Instantiate & fit clustering - this is done in one step for DBSCAN
dbs = DBSCAN(eps = 0.2, min_samples = min_samples).fit(circles)
```

```python
plt.scatter(circles[:,0], circles[:,1], c=dbs.labels_);
```

```python
db = DBSCAN(eps=0.4, min_samples=min_samples).fit(circles)
plt.scatter(circles[:,0], circles[:,1], c=db.labels_)
```
