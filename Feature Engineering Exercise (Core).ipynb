```python
import pandas as pd
```

```python
df= pd.read_csv('Data/bikeshare_train - bikeshare_train.csv')
df.head()
```

```python
df= df.drop(columns= ['casual','registered'])
df
```

```python
df['datetime']= pd.to_datetime(df['datetime'])
df.info()
```

```python
df['month_name']= df['datetime'].dt.month_name()
df['weekday']= df['datetime'].dt.day_name()
df['hour']= df['datetime'].dt.hour
df.head()
```

```python
df.dtypes
```

```python
df['hour']= df['hour'].astype(object)
```

```python
df= df.drop(columns=['datetime','season'])
df.head()
```

```python
def celsius_to_farenheit(celsius):
    return (celsius * 9/5) + 32
```

```python
df['temp'] = df['temp'].apply(lambda x: celsius_to_farenheit(x))

df['atemp'] = df['atemp'].apply(lambda x: celsius_to_farenheit(x))
df.head()
```

```python
def difference_in_temp(farenheit):
    if farenheit['temp']> farenheit['atemp']:
        return '+'
    else:
        return '-'
```

```python
df['temp_variance']= df.apply(difference_in_temp, axis=1)
df
```
